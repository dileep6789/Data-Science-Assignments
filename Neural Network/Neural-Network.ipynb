# Basic Neural Network

# create a structure

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
from sklearn.metrics import accuracy_score,classification_report
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV,KFold
```

# Create a Model

```python
def create_model():
    model=Sequential()
    model.add(Dense(16,input_dim=16,kernel_initializer="normal",activation='linear'))
    model.add(Dense(16,kernel_initializer="normal",activation='linear'))
    model.add(Dense(26,kernel_initializer="normal",activation='softmax'))

    adam=Adam()
    model.compile(loss="crossentropy",optimizer=adam,metrics=['accuracy'])
    return model
    
```

```python
model=KerasClassifier(model=create_model,verbose=1)
model.fit(train_x,train_y)
```

# Predictions

```python
y_pred=model.predict(test_x)
y_pred
```

```python
test_accuracy=accuracy_score(y_pred,test_y)*100
print("Accuracy:%.2f%%"%(test_accuracy))
```

# Batch Size and Epochs

```python
# build the structure
def create_model():
    model=Sequential()
    model.add(Dense(16,input_dim=16,kernel_initializer="normal",activation='linear'))
    model.add(Dense(16,kernel_initializer="normal",activation='linear'))
    model.add(Dense(26,kernel_initializer="normal",activation='softmax'))

    adam=Adam()
    model.compile(loss="crossentropy",optimizer=adam,metrics=['accuracy'])
    return model
```

# create a model

```python
model1=KerasClassifier(build_fn=create_model,verbose=1)
```

# Define parameters of Batch size and Epochs

```python
batch_size=[50,100,500,1000]
epochs=[10,50,100]
```

# Make in Dictionary as Params

```python
params=dict({"batch_size":batch_size,"epochs":epochs})
params
```

# Grid Search CV

```python
grid=GridSearchCV(estimator=model1,param_grid=params,verbose=1,cv=3,n_jobs=25)
grid_results1=grid.fit(train_x,train_y)
```

# Best Parameters and Score

```python
print(grid_results1.best_params_)
print(grid_results1.best_score_)
```

# Summarize the results

```python
print(f'Best{grid_results1.best_score_}, using {grid_results1.best_params_}')
means=grid_results1.cv_results_['mean_test_score']
std=grid_results1.cv_results_['std_test_score']
params=grid_results1.cv_results_['params']
for i,j,k in zip(means,std,params):
    print(f'{i} {j} using {k}')
```

# Dropout Rate and Learning Rate

```python
# Build the structure
from keras.layers import Dropout
```

```python
def create_model(rate,learning_rate):
    model=Sequential()
    model.add(Dense(16,input_dim=16,kernel_initializer="uniform",activation='softmax'))
    model.add(Dropout(rate))
    model.add(Dense(16,kernel_initializer="uniform",activation='softmax'))
    model.add(Dropout(rate))
    model.add(Dense(26,kernel_initializer="uniform",activation='softmax'))

    adam=Adam(learning_rate=learning_rate)
    model.compile(loss="crossentropy",optimizer=adam,metrics=['accuracy'])
    return model
```
