# Finding the Duplicated Rows

```python
data[data.duplicated(keep=False)]
```

# Removing the duplicated rows

```python
data1=data.drop_duplicates()
data1.reset_index(drop="index",inplace=True)
data1
```

# The rows with NaN Values

```python
data1[data1.isnull().any(axis=1)]
```

# If we impute this null rows the distribution changes so it is better to  remove

# Dropping the NaN values rows

```python
data2=data1.dropna()
data2.reset_index(drop="index",inplace=True)
data2
```

```python
data2.info()
```

```python
data2.describe()
```

```python
data2.isnull().sum()
```

# Width and MSTV are also.

```python
plt.figure(figsize=(12,8))
sns.heatmap(correlation_matrix,annot=True,linewidth=0.2,fmt='.3f')
```

# visualiztaions

```python
data2.columns
```

```python
plt.figure(figsize=(10,28))
for i,column in enumerate(data2.columns,1):
    plt.subplot(7,2,i)
    plt.suptitle("Histograms",y=1,color="blue",fontsize="18")
    sns.histplot(data=data2,x=data2[column],kde=True,color="peru")
    plt.xlabel(f"Histogram of {column}")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,28))
for i,column in enumerate(data2.columns,1):
    plt.subplot(7,2,i)
    plt.suptitle("Box plot",y=1,color="red",fontsize="18")
    sns.boxplot(data=data2,y=data2[column],color='yellow')
    plt.xlabel(f"Box plot of {column}")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,28))
for i,column in enumerate(data2.columns,1):
    plt.subplot(7,2,i)
    plt.suptitle("Violin and Box plot",y=1,color="red",fontsize="18")
    sns.boxplot(data=data2,y=data2[column],color='peru')
    sns.violinplot(data=data2,y=data2[column],color='yellow')
    plt.xlabel(f"{column}")

plt.tight_layout()
plt.show()
```
