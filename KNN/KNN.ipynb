# split data into X and Y

```python
X=data.iloc[:,1:-1]
X
```

```python
Y=data.iloc[:,-1]
Y
```

# Train and Test split

```python
from sklearn.model_selection import train_test_split
```

```python
train_x,test_x,train_y,test_y=train_test_split(X,Y,test_size=0.20,random_state=0)
```

```python
print("The Shape of Train X:",train_x.shape)
print("The Shape of Train Y:",train_y.shape)
print("The Shape of Test X:",test_x.shape)
print("The Shape of Test Y:",test_y.shape)
```

# Build the model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,classification_report
```

```python
model=KNeighborsClassifier(metric="euclidean") # by default n_neighbors=5
model.fit(train_x,train_y)
```

# Predictions

```python
train_pred=model.predict(train_x)
train_pred
```

```python
accuracy_score(train_pred,train_y)
```

```python
model.score(train_x,train_y)
```

```python
print(classification_report(train_pred,train_y))
```

# Training Accuracy:0.95

```python
y_pred=model.predict(test_x)
y_pred
```

```python
accuracy=accuracy_score(y_pred,test_y)*100
accuracy
```

```python
np.mean(y_pred==test_y)
```

```python
print(classification_report(y_pred,test_y))
```

# Testing Accuracy: 1.0

```python
model.n_neighbors
```

# Interview Questions

- What are the key hyperparameters in KNN?
    - The main hyperparameter to tune in k-nearest neighbors is k, the number of neighbors to consider. Other parameters include distance metrics, weights, and algorithm types.
    - By default n_neighbors(K)=5
    - other parameters are:
        - KNeighborsClassifier(n_neighbors=5, *, weights='uniform', algorithm='auto', leaf_size=30, p=2, metric='minkowski', metric_params=None, n_jobs=None)

- What distance metrics can be used in KNN?
    - Commonly used distance metrics for optimizing K-Nearest Neighbors (KNN) algorithms include Euclidean distance, Manhattan distance, Minkowski distance, and Cosine similarity.
    - By default metric = 'minkowski'
