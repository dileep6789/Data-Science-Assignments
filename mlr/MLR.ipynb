# converting into Dummy Variables

```python
car1=pd.get_dummies(car,dtype='int64',drop_first=True)
car1
```

```python
x1=car1.drop(columns=['Price'])
y1=car1['Price']
x1.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_x1,test_x1,train_y1,test_y1=train_test_split(x1,y1,test_size=0.2,random_state=11)
print(train_x1.shape,train_y1.shape)
```

```python
import statsmodels.api as sm
```

```python
model1=sm.OLS(train_y1,train_x1).fit()
model1.summary()
```

# here Automatic and Doors are having p value > 0.05

```python
car2=car1.drop(columns=['Automatic','Doors'])
car2
```

```python
x2=car2.drop(columns=['Price'])
y2=car2['Price']
```

```python
train_x2,test_x2,train_y2,test_y2=train_test_split(x2,y2,test_size=0.2,random_state=1)
print(train_x2.shape,train_y2.shape)
```

```python
model2=sm.OLS(train_y2,train_x2).fit()
model2.summary()
```

# Here Cyliders has only one Values
# remove the Drop Column

```python
car3=car2.drop(columns=['Cylinders'])
car3.head()
```

```python
x3=car3.drop(columns=['Price'])
y3=car3['Price']
```

```python
train_x3,test_x3,train_y3,test_y3=train_test_split(x3,y3,test_size=0.2,random_state=10)
print(train_x3.shape,train_y3.shape)
```

```python
model3=sm.OLS(train_y3,train_x3).fit()
model3.summary()
```

```python
car4=car3.drop(columns=['CC'])
car4
```

```python
x4=car4.drop(columns=['Price'])
y4=car4['Price']
```

```python
train_x4,test_x4,train_y4,test_y4=train_test_split(x4,y4,test_size=0.2,random_state=5)
print(train_x4.shape,train_y4.shape)
```

```python
model4=sm.OLS(train_y4,train_x4).fit()
model4.summary()
```

# Build the model with Entire Data

```python
X=car4.iloc[:,1:]
Y=car4.iloc[:,0]
```

```python
X
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=5)
```

```python
Linear=sm.OLS(Y,X).fit()
print(Linear.summary())
```

```python
Linear.fittedvalues
```

```python
plt.scatter(Y,Linear.fittedvalues)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
preds_train=Linear.predict(X_train)

mse_train=mean_squared_error(preds_train,Y_train)
print("The Mean Squared Error for training data :",mse_train)

rmse_train=np.sqrt(mse_train)
print("The Root Mean Squared Error for training data :",rmse_train)
```

```python
preds_test=Linear.predict(X_test)

mse_test=mean_squared_error(preds_test,Y_test)
print("The Mean Squared Error for testing data :",mse_test)

rmse_test=np.sqrt(mse_test)
print("The Root Mean Squared Error for testing data :",rmse_test)
```

# From the Final Model 
# R-Squared Value is 0.986
# Hence, It is good model 
# The Mean Square Error of testing data  is less than the training data..
# It is NOT Overfitting Model 

##  Lasso Regression Model

```python
from sklearn.linear_model import Lasso
```

```python
lasso=Lasso(alpha=0.2)
lasso.fit(X,Y)
```

```python
lasso_coeff=lasso.coef_
```

```python
lasso_coefficients=pd.DataFrame()
lasso_coefficients['columns']=X.columns
lasso_coefficients['Coefficient_Estimate']=pd.Series(lasso_coeff)
lasso_coefficients
```

```python
plt.figure(figsize=(10,8))
sns.barplot(x='Coefficient_Estimate',y='columns',data=lasso_coefficients)
```

## Ridge Regression Model

```python
from sklearn.linear_model import Ridge

ridge=Ridge(alpha=0.6)
ridge.fit(X,Y)
```

```python
ridge_coeff=ridge.coef_
```

```python
ridge_coefficients=pd.DataFrame()
ridge_coefficients['columns']=X.columns
ridge_coefficients['Coefficient_Estimate']=pd.Series(ridge_coeff)
ridge_coefficients
```

```python
plt.figure(figsize=(10,8))
sns.barplot(y='columns',x='Coefficient_Estimate',data=ridge_coefficients)
```

Both techniques are used to transform data into a common scale to make it easier to process and analyze.
- Standardization:
   Standardization is a data preprocessing technique used in statistics and machine learning to transform the features of your dataset so that they have a mean of 0 and a standard deviation of 1. This process involves rescaling the distribution of values so that the mean of observed values is aligned to 0 and the standard deviation to 1. 
   - Formula:
        	Z = (x-mean)/standard deviation
- Normalization:
   Normalization in machine learning is a data preprocessing technique used to change the value of the numerical column in the dataset to a common scale without distorting the differences in the range of values or losing information.
   - Formula:
            Xnorm = (X - Xmax) / (Xmax - Xmin)
