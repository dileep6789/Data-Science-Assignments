# Data Preprocessing

```python
data_train.isnull().sum()
```

# Imputing the Missing values

```python
data_train['Age'].fillna(data_train['Age'].median(),inplace=True) 

data_train['Embarked'].fillna(data_train['Embarked'].mode()[0],inplace=True)
```

# Since the missing values in Cabin column are 687 out of 891 imputing it will not help in better performance of the model.
# So we can drop the 'Cabin' column

```python
data_train.drop('Cabin',axis = 1,inplace=True)
```

```python
data_train.isna().sum()
```

# Encoding Categorical variables

```python
print("Unique categories in 'Sex' category are:")
data_train['Sex'].value_counts()
```

```python
print("Unique categories in 'Embarked' category are:")
data_train['Embarked'].value_counts()
```

```python
# Applying One-Hot encoding using dummy variables for the categories [Sex and Embarked]

data_train = pd.get_dummies(data = data_train,columns = ['Sex','Embarked'],drop_first=True,dtype='int64')
```

```python
print("Unique categories in 'Ticket' column are:",data_train['Ticket'].nunique())
print("Unique records in 'Name' column are :",data_train['Name'].nunique())
```

```python
# Since the columns [Ticket and Name] contains values a lot of distinct values we can drop these columns

data_train.drop(columns=['PassengerId','Name','Ticket'],inplace=True)
```

# Building Predictive Models
# Splitting the dataset into and test

```python
X=data_train.drop('Survived',axis=1)
X
```

```python
Y=data_train['Survived']
Y
```

```python
from sklearn.model_selection import train_test_split

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

# Splitting the data into training and testing dataset

```python
train_x,test_x,train_y,test_y=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
print('Shape of X Train is :',train_x.shape)
print('Shape of X Test is :',test_x.shape)
print('Shape of Y Train is :',train_y.shape)
print('Shape of Y Test is :',test_y.shape)
```

# Light BGM

```python
import lightgbm as lgb
```

# Building the model

```python
LGB_classifier=lgb.LGBMClassifier()
```

# Training it using train data

```python
LGB_classifier.fit(train_x,train_y)
```

# Making predictions using LGBM Classifier

```python
LGB_preds=LGB_classifier.predict(test_x)
LGB_preds
```

# Evaluating the model on the testing data

```python
accuracy_LGBM=accuracy_score(LGB_preds,test_y)
print('The Accuracy of LGBM Algorithm is:',accuracy_LGBM)
```

```python
precision_LGBM = precision_score(LGB_preds,test_y)

print('The Precision score of LGBM Algorithm is:',precision_LGBM)
```

```python
recall_LGBM = recall_score(LGB_preds,test_y)

print('The Recall score of LGBM Algorithm is:',recall_LGBM)
```

```python
f1_LGBM = f1_score(LGB_preds,test_y)

print('The F1 Score of LBGM Algorithm is:',f1_LGBM)
```

```python
print('Classification Report for LGBM is:\n')
print(classification_report(test_y,LGB_preds))
```

# Using Cross-validation technique and Hyper parameter Tuning

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.model_selection import GridSearchCV
# using Grid-search cv for hyperparameter Tuning
parameters={
    'learning_rate':[0.1,0.01],
    'num_leaves':[10,20],
    'max_depth':[1,2,3],
    'n_estimators':[100,200],
    'min_split_gain':[0.1,0.3],
    'min_child_samples':[10,15,20]
}
parameters
```

```python
kfold=KFold()
results=cross_val_score(LGB_classifier,train_x,train_y,cv=kfold)
```

```python
results.mean()
```

# Grid Search CV

```python
grid1=GridSearchCV(estimator=LGB_classifier,param_grid=parameters,cv=kfold,n_jobs=5)
grid1.fit(train_x,train_y)
```

```python
print('The best score obtained is:',grid1.best_score_,'\n')

print('The best parameters are:')
grid1.best_params_
```

```python
RESULTS.mean()
```

# Using Grid-SearchCV for hyperparameter tuning

```python

grid2 = GridSearchCV(estimator=XGB_classifier,param_grid=Hyperparameters,cv=KFOLD)

grid2.fit(train_x,train_y)
```

```python
print('The best score obtained is:',grid2.best_score_,'\n')

print('The best parameters are:')
grid2.best_params_
```

# Compartive Analysis

```python
dict={
    'LGBM':[accuracy_LGBM,precision_LGBM,recall_LGBM,f1_LGBM],
    'XGBM':[accuracy_XGBM,precision_XGBM,recall_XGBM,f1_XGBM]
}
comparsion=pd.DataFrame(dict,index=['Accuracy','Precision','Recall','F1-Score'])
comparsion
```

From the above visualization we can see that XG Boost has slight better peformance metrics like accuracy,precision,F1-Score and Light GB has little bit high recall when compared to XGBoost on this dataset.
