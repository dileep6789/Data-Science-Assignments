# Recommendation

```python
anime_genre=anime.pivot(index="anime_id",
                       columns="genre",
                       values="rating")
anime_genre
```

```python
anime_genre.shape
```

```python
anime_genre.head()
```

```python
anime_genre.notnull().sum().sum()
```

```python
anime_genre.isnull().sum().sum()
```

```python
anime_genre.tail()
```

```python
anime_genre.fillna(0,inplace=True)
anime_genre
```

```python
from sklearn.metrics import pairwise_distances
```

```python
anime_sim=1-pairwise_distances(anime_genre,metric="cosine")
anime_sim
```

```python
anime_sim_df=pd.DataFrame(anime_sim)
anime_sim_df
```

```python
np.fill_diagonal(anime_sim,0)
anime_sim_df
```

# 1.Can you explain the difference between user-based and item-based collaborative filtering?

User-based and item-based collaborative filtering are both methods used in recommendation systems to predict user preferences, but they differ in how they generate recommendations.

## User-Based Collaborative Filtering
User-Based Collaborative Filtering recommends items to a user based on the preferences of similar users. It identifies users with similar behaviors and suggests items that those users liked but the target user hasn’t interacted with yet. While it provides personalized recommendations, it can struggle with new users (cold-start problem) and requires more computation when there are many users.

## Item-Based Collaborative Filtering
Item-Based Collaborative Filtering recommends items based on their similarity to items the user has interacted with. It looks at item similarities (based on user ratings or interactions) and suggests similar items. This method is more scalable and handles new users better, but it may be less personalized compared to user-based filtering.

# 2.What is collaborative filtering, and how does it work?


Collaborative filtering is a recommendation technique used to predict a user's preferences by analyzing the preferences of similar users or items. It assumes that users who have agreed on certain items in the past will likely agree on others in the future.

## How it works:
*  Data Collection: It gathers data on user interactions, such as ratings, purchases, or browsing history.
*  Similarity Calculation: Based on this data, it identifies similarities between users (user-based) or items (item-based) using methods like cosine similarity or Pearson correlation.
*  Recommendation Generation: Once similarities are established, the system recommends items that similar users have liked or items similar to those the user has interacted with.
* Collaborative filtering is commonly used in e-commerce, streaming services, and social media platforms to personalize user experiences.
