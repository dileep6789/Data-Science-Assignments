# Additional Notes

1. Explain Bagging and Boosting methods. How is it different from each other.

### Bagging:

- Bagging stands for **Bootstrap Aggregating.**

- It is an ensemble learning technique which uses Bootstrap Aggregation method to build multiple models on random samples and group them together to build a stronger model.

- Examples for bagging are : Bagging Classifier, Random Forest Classifier.


### Boosting:

- Boosting is also an ensemble learning technique which builds multiple models and group them together to form a strong model.

- Some of the examples for Boosting methods are : Adaboost, Gradient Boosting etc.

*The difference between Bagging and Boosting is as follows:*

- Bagging uses bootstrap method for sampling i.e. it picks random samples from the data (with replacement) and build models on them and ensemble (combine) them all together at the end to create a strong learning model.

- In Boosting, the sampling does not take place randomly, but the sampling will be done based on the results of previous models.
  First random model is created and if there is any error the sampling will happen again in order to correct the previous error and so on until maximum accuracy / minimum error is achieved.

- In Bagging, the models are built in a parallel manner. That is multiple individual models are built simultaneously and in the end they are grouped together to form a strong model.

- Whereas in Boosting, the models are built one by one, based on previous models.

2.Explain how to handle imbalance in the data.

- Imbalance in the data refers to uneven distribution of classes in the target column in the data. Means when one class has higher records and some other class has lower records when compared to other. So this creates bias in the models towards the class which are in majority

- In our case also, the class Type-2 had 76 observations whereas the Type-6 had only 6 records. So this will effect in more bias towards the class Type-2. So in order to avoid these kind of problems it is necessary to handle the imbalance in the data.


- *The imbalance in the data can be done by following methods:*

   - **Evaluation Metric :** Choosing the appropriate evaluation metric becomes important when dealing with imbalance data because a high accuracy always does not mean that our model is performing well. We also need to consider other evaluation metrics like: Precision, Recall and F1 score
   
   - **Sampling methods :** Performing different sampling methods result in adjusting the imbalance. For example we can use undersampling for the classes which are high in number and perform oversampling to create classes which were fewer in number inorder to have that balance between them.
   
   - **Ensemble Techniques :** Using different ensemble techniques like bagging and boosting helps us to handle the imbalance in the data
